```python
import pandas as pd
```

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'grade': ['Senior', 'Junior', 'Middle', 'Senior', 'Junior',
              'Middle', 'Senior', 'Senior', 'Middle', 'Junior'],
    'location': ['RU', 'AU', 'US', 'RU', 'JP', 'RU', 'US', 'US', 'RU', 'JP'],
    'salary': [700, 45, 96, 750, 50, 120, 800, 7200, 180, 71],
    'hours_worked': [16, 10, 15, 65, 40, 20, 70, 175, 50, 145]
}

data = pd.DataFrame(data)

data
```

Задача №1.

Создать новый столбец с категориями зарплат, используя сформулированные критерии:
1. Если зарплата меньше 50, присвоить категорию A.
2. Если зарплата больше 50, но меньше 100, присвоить категорию B.
3. Если зарплата находится в диапазоне от 100 до 500, присвоить категорию C.
4. Для всех остальных значений назначить категорию Other.

```python
data['salary_category'] = 'other'
data
```

```python
data.loc[data['salary'] <= 50, 'salary_category'] = 'A'
data
```

```python
data.loc[(data['salary'] > 50) & (data['salary'] <= 100), 'salary_category'] = 'B'
data.loc[(data.salary > 100) & (data.salary <= 500), 'salary_category'] = 'C'
```

```python
data
```

```python
bins = [-float('inf'), 50, 100, 500, float('inf')]
labels = ['A', 'B', 'C', 'other']
data['salary_category_cut'] = pd.cut(data['salary'], bins=bins, labels=labels, right=False)
data
```

Задача №2.

Необходимо вывести по два сотрудника с наибольшим количеством отработанных часов для каждого уровня квалификации (грейда).

```python
data_1 = data.groupby('grade').agg({'hours_worked': 'sum'}).reset_index()
data_1.columns = ['grade', 'total_hours_worked']
data_1
```

```python
top_workers_per_grade = data.groupby('grade').apply(lambda x: x.nlargest(2, 'hours_worked')).reset_index(drop=True)

top_workers_per_grade
```

```python
data_1_sorted = data.sort_values(by=['grade', 'hours_worked'], ascending=[True, False])
data_1_sorted
```
